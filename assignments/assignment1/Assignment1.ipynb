```python
# TASK 1
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# a) Filter out invalid transactions where the price is zero or the quantity is negative
valid_transactions = [
    t for t in transactions
    if t["price"] > 0 and t["quantity"] >= 0
]

print(valid_transactions)
```

```python
# b) For each remaining transaction, compute a new field total = price * quantity.
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

print(valid_transactions)
```

```python
# c) Group the valid transactions by region and compute the total revenue per region. Store the result in a dictionary.
revenue_by_region = {}

for t in valid_transactions:
    region = t["region"]
    total = t["total"]
    
    if region not in revenue_by_region:
        revenue_by_region[region] = 0
    
    revenue_by_region[region] += total

print(revenue_by_region)
```

```python
# Print the region with the highest total revenue and its value using an f-string.
max_region = max(revenue_by_region, key=revenue_by_region.get)
max_value = revenue_by_region[max_region]

print(f"Region with highest revenue: {max_region} ({max_value})")
```

```python
# TASK 2

# Call the function with the list above and print the returned dictionary.
def summarise_sales(sales):
    total = sum(sales)
    average = round(total / len(sales), 2)
    
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1
    
    above_average_count = sum(1 for s in sales if s > average)
    
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }


sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

result = summarise_sales(sales)
print(result)
```

```python
# TASK 3

#  a) Create a new list called passing that contains only the names of students with a grade of 50 or above. Print the list.

students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

passing = []

for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print(passing)
```

```python
# b) Loop through the students and print a status message for each. If the student's grade is 80 or above, 
# print "[name]: Excellent"; if between 50 and 79, print "[name]: Pass"; otherwise print "[name]: Fail".

for student in students:
    name = student["name"]
    grade = student["grade"]
    
    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")
```

```python
# c) Count how many students have an attendance rate below 0.80. Print the count.count = 0
count = 0

for student in students:
    if student["attendance"] < 0.80:
        count += 1

print(count)
```

```python
# d) Find the student with the highest grade using a loop (do not use max()). Print the student's name and grade.

best_student = None

for student in students:
    if best_student is None or student["grade"] > best_student["grade"]:
        best_student = student

print(f"{best_student['name']}: {best_student['grade']}")
```

```python
# e) Now find the student with the highest grade using max(). Print the student's name and grade.
best_student = max(students, key=lambda student: student["grade"])

print(f"{best_student['name']}: {best_student['grade']}")
```

```python
# TASK 4

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # stores assets: ticker -> {shares, price_per_share}

    # a) Add a new asset or update existing one
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings:
            # update shares and price if asset exists
            self.holdings[ticker]["shares"] += shares
            self.holdings[ticker]["price_per_share"] = price_per_share
        else:
            # add new asset
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

    # b) Calculate total value of portfolio
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            total += shares * price  # value of each asset
        return total

    # c) Find the most valuable asset
    def most_valuable(self):
        best_ticker = None
        best_value = 0

        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            value = shares * price

            # update if this asset has higher value
            if best_ticker is None or value > best_value:
                best_ticker = ticker
                best_value = value

        return best_ticker

    # d) Return readable summary of portfolio
    def __str__(self):
        return f"Portfolio of {self.owner}: {len(self.holdings)} assets, total value = {self.total_value()}"
                
```

```python
# WE test to see if it worked by testing some random numbers
portfolio = Portfolio("Alice")

portfolio.add_asset("AAPL", 10, 150)
portfolio.add_asset("GOOGL", 5, 2000)
portfolio.add_asset("TSLA", 8, 700)

print("Total value:", portfolio.total_value())
print("Most valuable:", portfolio.most_valuable())
print(portfolio)
```

```python
# Task 5
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

dates = []

# convert strings to date objects
for order in orders:
    date_obj = datetime.strptime(order[0], "%Y-%m-%d").date()
    dates.append(date_obj)

# find earliest and latest
earliest = min(dates)
latest = max(dates)

# calculate difference
difference = (latest - earliest).days

print(difference)
```

```python
# Task 7

import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

df = pd.DataFrame(data)

# a) salary per year of experience
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round()
print("a)\n", df)

# b) top 3 by performance score
top3 = df.sort_values(by="performance_score", ascending=False).head(3)
print("b)\n", top3)

# c) Finance employees with score > 4.0
finance_high = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0)
]
print("c)\n", finance_high)

# d) rename column
df = df.rename(columns={"years_exp": "experience_years"})
print("d)\n", df)

# e) select salary and performance for first 4 employees
subset = df.iloc[:4][["salary", "performance_score"]]
print("e)\n", subset)
```
